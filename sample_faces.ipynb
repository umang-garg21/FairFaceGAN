```python
from matplotlib import pyplot as plt
import random
import numpy as np
from PIL import Image
```

```python
IMG_DIR = './data/imgs_jpg/'
```

```python
max_ind = 749

rand_ind = random.choices(np.arange(max_ind+1), k=5)
```

```python
races = ["east_asian", "indian", "black", "white", "middle_eastern", "latino_hispanic", "southeast_asian"]
```

```python
def get_img_paths(race):

    img_paths = []

    for ind in rand_ind:

        img_path = ''
        img_path += IMG_DIR + race + f'/{race}_{ind}.jpg'
        img_paths.append(img_path)

    return img_paths
```

```python
img_paths_black = get_img_paths('black')
img_paths_ea = get_img_paths('east_asian')
img_paths_indian = get_img_paths('indian')
img_paths_lh = get_img_paths('latino_hispanic')
img_paths_me = get_img_paths('middle_eastern')
img_paths_sa = get_img_paths('southeast_asian')
img_paths_w = get_img_paths('white')
```

```python
races_img_paths = [img_paths_black, img_paths_ea, img_paths_indian, img_paths_lh, img_paths_me, img_paths_sa, img_paths_w]
```

```python
fig, axs = plt.subplots(7, 5, sharex=True, sharey=True, figsize=(10, 10))

plt.subplots_adjust(wspace=0, hspace=0)

for ax_row, race_imgs in zip(axs, races_img_paths):
    for ax, img_path in zip(ax_row, race_imgs):
        img = Image.open(img_path)
        ax.axis('off')
        ax.imshow(img)

plt.savefig('./data/plots/sample_faces.png')
```
